```python
import tkinter as tk
from tkinter import messagebox
import cv2
import os
from PIL import Image
import numpy as np
from tkinter import *
```

```python
window = tk.Tk()
window.title("Student attendance system")
window.configure(bg='#F2E5E5')

l1 = tk.Label(window, text="Name")
l1.grid(column=0,row=0)
f1 = tk.Entry(window,width=50,bd=2)
f1.grid(column=1,row=0)

l2 = tk.Label(window, text="Surname")
l2.grid(column=0,row=1)
f2 = tk.Entry(window,width=50,bd=2)
f2.grid(column=1,row=1)

l3 = tk.Label(window, text="Age")
l3.grid(column=0,row=2)
f3 = tk.Entry(window,width=50,bd=2)
f3.grid(column=1,row=2)

def generate_dataset():
    # Cascade classifier
    face_classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
    # convert image in grayscale image from rgb
    def face_cropped(img):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray,1.2,5)

        if faces is ():
            return None
        for(x,y,w,h) in faces:
            cropped_face=img[y:y+h,x:x+w]
        return cropped_face 
    camera = cv2.VideoCapture(0)
    id=2
    img_id=0

    while True:
        ret,frame = camera.read()
        if face_cropped(frame) is not None:
            img_id+=1
            face = cv2.resize(face_cropped(frame),(300,300))
            face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
            file_path = "Data/user."+str(id) +"."+str(img_id)+".jpg"
            cv2.imwrite(file_path,face)
            cv2.putText(face,str(img_id),(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
            cv2.imshow("Cropped face", face)
            if cv2.waitKey(1)==13 or int(img_id)==200:
                break
    camera.release()
    cv2.destroyAllWindows()
    print("Collecting samples is completed...............")

b1 = tk.Button(window, text="Gjenero datasetin", bg='#CE7777',fg='#2B3A55', command=generate_dataset)
b1.grid(column=0, row=4)

b2 = tk.Button(window, text="Merr vijueshmerine", bg='#CE7777',fg='#2B3A55')
b2.grid(column=1, row=4)

b3 = tk.Button(window, text="Train", bg='#CE7777',fg='#2B3A55')
b3.grid(column=2, row=4)

window.geometry("800x500")
window.mainloop()
```
